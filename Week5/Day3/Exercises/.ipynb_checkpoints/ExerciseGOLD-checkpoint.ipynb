```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.DataFrame({
        'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
        'Temperature': [22, 24, 23, 24, 25, 26, 25],
        'Rainfall': [1, 0, 2, 1, 0, 0, 1]
    })

plt.plot(df['Day'], df['Temperature'], label='Temp', color='tab:blue')
plt.errorbar(df['Day'], df['Temperature'], yerr=1)
plt.plot(df['Day'], df['Rainfall'], label='Rain', color='tab:orange')
plt.errorbar(df['Day'], df['Rainfall'], yerr=1)
 
```
