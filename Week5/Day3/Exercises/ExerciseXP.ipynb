```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]  # Prices
y = [5, 4, 3, 2, 1]  # Demand

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Product Demand vs Price")
ax.set_ylabel("Demand")
ax.set_xlabel("Price")
```

```python
x = [1, 2, 3, 4, 5]  # Age
y = [75, 80, 88, 95, 105]  # Height in cm

fig, ax = plt.subplots()
ax.scatter(x, y)
ax.set_title("Childrens Height vs Age")
ax.set_ylabel("Height (cm)")
ax.set_xlabel("Age (years)")
```

```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(1, 2, figsize=(6, 4))
ax[0].plot(x, y1, 'tab:blue')
ax[0].set_title('y = sin(x)')

ax[1].scatter(x, y2, marker='s')
ax[1].set_title('y = cos(x)')
```

```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(2, 1)
ax[0].plot(x, y1, label='y = sin(x)')
ax[0].legend()
ax[1].scatter(x, y2, label='y = cos(x)')
ax[1].legend()
```
