```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,10) #Array of range 0 to 9
y=x**3
print(x)
print(y)
plt.plot(x,y)
plt.title('Line Chart')
plt.xlabel('X-Axis') 
plt.ylabel('Y-Axis')
```

```python
x = np.arange(0, 11)
y = x ** 3
x2 = np.arange(0, 11)
y2 = (x ** 2) / 2
plt.plot(x,y,color='r',label='first data', linewidth=5) 
plt.plot(x2,y2,color='y',linewidth=5,label='second data')
plt.title('Multiline Chart')

# Uses the label attribute to display reference in legend
plt.ylabel('Y axis')
plt.xlabel('X axis')

# Shows the legend in the best postion with respect to the graph
plt.legend()
```

```python
x = ["India",'USA',"Japan",'Australia','Italy']
y = [6,7,8,9,2]

# Printing the variables
print(x)
print(y)

plt.bar(x,y, label='Bars1', color ='r') # Function to plot

# Function to give x and y labels 
plt.xlabel("Country")
plt.ylabel("Inflation Rate%")

# Function to give heading of the chart
plt.title("Bar Graph")
```

```python
# Generation of 1 set of variables 
x = ["India",'USA',"Japan",'Australia','Italy']
y = [6,7,8,9,5]

# Generation of 2 set of variables
x2 = ["India",'USA',"Japan",'Australia','Italy']
y2 = [5,1,3,4,2]

# Printing all variables
print(x,y,x2,y2,sep="\n")

# Functions to plot 
plt.bar(x,y, label='Inflation', color ='y')
plt.bar(x2,y2, label='Growth', color ='g')

# Functions to give x and y labels
plt.xlabel("Country")
plt.ylabel("Inflation & Growth Rate%")

plt.title("Multiple Bar Graph")
plt.legend()
```

```python
raw_data={'names':['Nick','Sani','John','Rubi','Maya'],
'jan_score':[123,124,125,126,128],
'feb_score':[23,24,25,27,29],
'march_score':[3,5,7,6,9]}

# Segregating the raw data into usuable form/variables
df=pd.DataFrame(raw_data,columns=['names','jan_score','feb_score','march_score'])
df['total_score']=df['jan_score']+df['feb_score']+df['march_score']

# Printing the data
print(df)

# Function to plot the graph
plt.pie(df['total_score'],labels=df['names'],autopct='%.2f%%')
plt.axis('equal')
plt.axis('equal')
```

```python
plt.figure(figsize=(10,10))

# Generation of variables
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([5,2,4,2,1,4,5,2])

# Generating 4 subplots in form of 2x2 matrix
# In the line below the arguments of plt.subplot are as follows:
# 2- no. of rows
# 2- no. of columns
# 1- position in matrix
# Position (0,0)
plt.subplot(2,2,1)
plt.plot(x,y,'g')
plt.title('Sub Plot 1')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# Position (0,1)
plt.subplot(2,2,2)
plt.plot(y,x,'b')
plt.title('Sub Plot 2')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# Position (1,0)
plt.subplot(2,2,3)
plt.plot(y*2,x*2,'y')
plt.title('Sub Plot 3')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# Position (1,1)
plt.subplot(2,2,4)
plt.plot(x*2,y*2,'m')
plt.title('Sub Plot 4')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

# Function for layout and spacing
plt.tight_layout(h_pad=5, w_pad=10)
```

```python
x = np.linspace(0, 10, 20)
y1 = x ** 2
y2 = x ** 1.5

plt.subplot(1, 2, 1) # One row, two columns, first plot
plt.plot(x, y1, color='red', marker='o', label='y = x**2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Plot of y1 against x')
plt.legend()

plt.subplot(1, 2, 2) # Second plot
plt.scatter(x, y2, color='blue', marker='s', label='y = x**1.5')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot of y2 against x')
plt.legend()

plt.tight_layout()
```

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

fig, axs = plt.subplots(2, 2, figsize=(12, 8))

axs[0, 0].plot(x, y, color='blue', label='y = sin(x^2)')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
max_y = np.max(y)
min_y = np.min(y)
axs[0, 0].annotate(f'Max: {max_y:.2f}', xy=(x[np.argmax(y)], max_y), xytext=(10, 10),
                   textcoords='offset points', color='red')
axs[0, 0].annotate(f'Min: {min_y:.2f}', xy=(x[np.argmin(y)], min_y), xytext=(-40, -20),
                   textcoords='offset points', color='red')
axs[0, 0].legend(title='Legend', loc='upper right', shadow=True)

axs[0, 1].plot(x, -y, color='orange', label='y = -sin(x^2)')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('-y')
max_neg_y = -np.min(y)
min_neg_y = -np.max(y)
axs[0, 1].annotate(f'Max: {max_neg_y:.2f}', xy=(x[np.argmin(y)], -min_y), xytext=(10, 10),
                   textcoords='offset points', color='red')
axs[0, 1].annotate(f'Min: {min_neg_y:.2f}', xy=(x[np.argmax(y)], -max_y), xytext=(-40, -20),
                   textcoords='offset points', color='red')
axs[0, 1].legend(title='Legend', loc='upper right', shadow=True)

axs[1, 0].plot(x, y ** 2, color='green', label='y^2 = sin(x^2)')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y^2')
max_y2 = np.max(y ** 2)
min_y2 = np.min(y ** 2)
axs[1, 0].annotate(f'Max: {max_y2:.2f}', xy=(x[np.argmax(y ** 2)], max_y2), xytext=(10, 10),
                   textcoords='offset points', color='red')
axs[1, 0].annotate(f'Min: {min_y2:.2f}', xy=(x[np.argmin(y ** 2)], min_y2), xytext=(-40, -20),
                   textcoords='offset points', color='red')
axs[1, 0].legend(title='Legend', loc='upper right', shadow=True)

axs[1, 1].plot(x, -(y ** 2), color='red', label='-y^2 = -sin(x^2)')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('-y^2')
max_neg_y2 = -np.min(y ** 2)
min_neg_y2 = -np.max(y ** 2)
axs[1, 1].annotate(f'Max: {max_neg_y2:.2f}', xy=(x[np.argmin(y ** 2)], -min_y2), xytext=(10, 10),
                   textcoords='offset points', color='red')
axs[1, 1].annotate(f'Min: {min_neg_y2:.2f}', xy=(x[np.argmax(y ** 2)], -max_y2), xytext=(-40, -20),
                   textcoords='offset points', color='red')
axs[1, 1].legend(title='Legend', loc='upper right', shadow=True)

plt.tight_layout()
```
