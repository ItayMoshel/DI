```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
```

```python
boston_data = fetch_openml(data_id=506, parser='auto')
df = pd.DataFrame(data=boston_data.data, columns=boston_data.feature_names)
df['MEDV'] = boston_data.target

print(df.describe(), '\n\n\n\n\n')
print(df.head(), '\n\n\n\n\n')
print(df.info())
```

```python
plt.figure(figsize=(12, 8))
df.hist(bins=20, edgecolor='black')
plt.tight_layout()
```

```python
len_features = len(df.columns)
fig, axs = plt.subplots(len_features // 3, 3, figsize=(12, 8))
i = 0
for feature in df.columns:
    if feature == "MEDV":
        continue
    row, col = i//3, i%3
    ax = axs[row, col]
    ax.scatter(df[feature], df['MEDV'])
    # ax.set_title(f"{feature} vs MEDV")
    if col == 0:
        ax.set_ylabel('MEDV')
    ax.set_xlabel(feature)
    i += 1

plt.setp(axs[:, 0], ylabel='MEDV')
plt.tight_layout()
```

```python
missing_data = df.isnull().sum()
print(f"Missing Data: {missing_data}")
```

```python
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
print("\nNormalized Data:\n", df_scaled.head())
```

```python
bins = [0, 25, 50, 100]
labels = ['new', 'old', 'very old']
df['AGE_category'] = pd.cut(df['Age'], bins=bins, labels=labels)
encoder = OneHotEncoder(drop='first', sparse_output=False)  # Update parameter name
age_encoded = encoder.fit_transform(df[['AGE_category']])
age_encoded_df = pd.DataFrame(age_encoded, columns=[f'AGE_{label}' for label in labels[:-1]])
df_encoded = pd.concat([df.drop('AGE_category', axis=1), age_encoded_df], axis=1)

print(df_encoded.head())
```

```python
label_mapping = {label: index for index, label in enumerate(labels)}
df['AGE_category'] = df['AGE_category'].map(label_mapping)
correlation_matrix = df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
```

#### 

```python
new_properties = df[df['Age'] <= 35]
old_properties = df[(df['Age'] > 35) & (df['Age'] <= 70)]
very_old_properties = df[df['Age'] > 70]

plt.figure(figsize=(10, 6))
plt.scatter(new_properties['Age'], new_properties['MEDV'], label='New Properties')
plt.scatter(old_properties['Age'], old_properties['MEDV'], label='Old Properties')
plt.scatter(very_old_properties['Age'], very_old_properties['MEDV'], label='Very Old Properties')

plt.xlabel('Property Age')
plt.ylabel('Price (MEDV)')
plt.title('Relationship between Property Age and Price')
plt.legend()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Property Age')
plt.ylabel('Frequency')
plt.title('Histogram of Property Age')

plt.figure(figsize=(10, 6))
plt.hist(df['MEDV'], bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Median Value (Price)')
plt.ylabel('Frequency')
plt.title('Histogram of Median Value')
```
